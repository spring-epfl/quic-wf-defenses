```python
# percentages of Google-owner subresources
import glob
import json
import numpy as np
from urllib.parse import urlparse
from dateutil import parser

owners_mapping = np.load('../tracker-radar/domain-owner-mapping.npy', allow_pickle=True).item()
owners_mapping_our_dataset = dict()


def is_google_domain(d):
    d = d.lower().strip()

    if d.endswith('ggpht.com'):
        return True
    if "google" in d or "youtube" in d or "doubleclick" in d or "gstatic.com" in d:
        return True

    return False

def har_to_connections(file='cluster-150-3/hars/thetxthub.com.har'):
    global owners_mapping_our_dataset

    try:
        data_har = []
        with open(file) as f:
            data_har = json.load(f)

        subresources = []
        t0 = None
        for row in data_har:
            # simplify URL into domain
            url = row[0]
            domain = urlparse(url).netloc

            subresources.append(domain)


        return subresources
    except:
        return []

folder = 'quic-100p-338'
hars = glob.glob(folder+"/hars/*")

count = 0
for h in hars:
    url = h.replace(folder+"/hars/", '').replace('.har', '')
    sr = list(set(har_to_connections(h)))
    sr_google = [x for x in sr if is_google_domain(x)]
    print(url, sr, sr_google)
    if len(sr_google) > 0:
        count += 1

print("Total website that use a google resource", count)
```
